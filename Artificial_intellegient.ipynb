### **Histogram**

```python
for i in df.select_dtypes(include='number').columns:
  sns.histplot(data=df,x=i)
  plt.show()
```

```python
for i in df.select_dtypes(include='number').columns:
  sns.boxplot(data=df,x=i)
  plt.show()
```

```python
df.select_dtypes(include='number').columns
```

```python
for i in ['Year', 'Adult Mortality', 'infant deaths',
       'Alcohol', 'percentage expenditure', 'Hepatitis B', 'Measles ', ' BMI ',
       'under-five deaths ', 'Polio', 'Total expenditure', 'Diphtheria ',
       ' HIV/AIDS', 'GDP', 'Population', ' thinness  1-19 years',
       ' thinness 5-9 years', 'Income composition of resources', 'Schooling']:
       sns.scatterplot(data=df,x=i,y='Life expectancy ')
       plt.show()
```

```python
coo=df.select_dtypes(include='number').corr()
sns.heatmap(coo,annot=True)
```

```python
plt.figure(figsize=(12,12))
sns.heatmap(coo,annot=True)
```

```python
for col in df.select_dtypes(include='number').columns:
  data=df[col].value_counts()
  plt.figure(figsize=(6,6))
  plt.pie(data,labels=data.index,autopct='%.2f%%',startangel=90)
  plt.title(f"pie chart for {col}")
  plt.axis('equal')
  plt.show()
```
