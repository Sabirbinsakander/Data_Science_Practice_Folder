```python
import pandas as pd
import numpy as np
# import matplotlib.pyplot as plt
# import seaborn as sns
```

```python
df = pd.read_csv('/content/sample_data/Mall_Customers.csv')
df.head(5)
```

```python
x = df.iloc[:, -2:].values
x[0], x.shape
```

```python
from sklearn.cluster import KMeans

ssd = []
for k in range(1, 11):
  k_means = KMeans(n_clusters=k, random_state=39)
  k_means.fit(x)
  ssd.append(k_means.inertia_)
```

```python
import matplotlib.pyplot as plt

plt.title("The Elbow Method")
plt.xlabel("k")
plt.ylabel("SSD")
plt.plot(range(1, 11),ssd)
plt.show()
```

```python
k_means = KMeans(n_clusters=5, random_state=39)
Y = k_means.fit_predict(x)
Y.shape
```

```python
Y[0]
```

```python
data = df.iloc[:, -2:]
data['cluster'] = Y
data.head(5)
```
